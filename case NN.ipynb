```python
# import the correct library for using the join
import pandas as pd
import numpy as np
```

```python
facility_crm = pd.read_excel('case_study.xlsx', sheet_name='facility_crm')
crm = pd.read_excel('case_study.xlsx', sheet_name='CRM')
facility = pd.read_excel('case_study.xlsx', sheet_name='facility')
```

##### Question for easy to read. 

For each position_dt in **facility** find the **percentage** of **original_account_id** & **position_dt** where **valuation_dt** in CRM is 3 years older than **position_dt**.    

hint: join facility with facility crm on original_account_id and keep the relevant attributes. Join the resulted table with crm on position_dt and cover_id and keep the relevant attributes.

```python
# Check the collumn of crm to easy to see the collum inside the notebook without looking to excel
crm.columns.values.tolist()
```

```python
# Check the collumn of crm to easy to see the collum inside the notebook without looking to excel
facility.columns.values.tolist()
```

```python
# reason to join according to what i understand from the hind is using orginial_account_id. However some situation we can join two same collumn with datetime. 
first_df_merge = facility_crm.merge(facility, on=['original_account_id','position_dt'])
# for this solution i would choose join only original_account_id, and keep other collumns.
# first solution to join. ( direct from table to table )
second_df_merge = pd.merge(facility_crm, facility,  how="left", on=['original_account_id'])
# first solution to join. ( pandas join two table, we can also added function left_join, how="right", depend on purpose)
```

```python
first_df_merge
```

```python
lastjoindf = pd.merge(first_df_merge,crm, on=['cover_id', 'position_dt'])
# join both collumns on two key cover_id, position_id
```

```python
lastjoindf['diff_years'] = (lastjoindf['position_dt'] - lastjoindf['valuation_dt']) / np.timedelta64(1, 'Y')
lastjoindf['diff_years'] = round(lastjoindf['diff_years'])
## calculate the different year between position_df and valuation_dt to find out if valuation_dt is older than position_df 3 years. 
```

```python
lastjoindf
```

For each position_dt in facility find the percentage of original_account_id & position_dt where valuation_dt in CRM is 3 years older than position_dt. 
According to what i observed from the question is pairs of both collumn original_account & position_dt collumn. So i made the next script to calculate the number of value of both collumns. ( it might have another way of understanding the question )

```python
count = lastjoindf.groupby(["original_account_id", "position_dt"]).size().reset_index(name="count")
count = count.shape[0]
# Print the result
print("Number of unique original_account_id & position_dt pairs:", count)
```

```python
count_diff = lastjoindf[lastjoindf["diff_years"] > 3].groupby(["original_account_id","position_dt"]).size().reset_index(name="count")
count_diff = count_diff.shape[0]
print("Number of original_account_id & position_dt pairs which valuation_dt is older than position_dt 3 year:", count_diff)
```

```python
print("Percentage of original_account_id & position_dt pairs which valuation_dt is older than position_dt 3 year :", round((count_diff/count)*100),  '%%')
```
